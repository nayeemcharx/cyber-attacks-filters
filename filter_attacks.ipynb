```python

import json
import re
import pandas as pd

# Function to extract the "request" field if the search string is found
def extract_request(log_file_path,search_string,output_file_path):
    extracted_requests = []
    # Define the regular expression pattern to check for the string
    search_pattern = re.compile(rf"{search_string}",re.IGNORECASE)
    with open(log_file_path, "r") as file:
        for line in file:
            try:
                # Parse the JSON object from the line
                log_entry = json.loads(line)
                
                # Check if the specific string is in the log entry
                if search_pattern.search(line):
                    # Extract and print the "request" key if the string is found
                    if "request" in log_entry:
                        print(log_entry["request"])
                        extracted_requests.append(log_entry["request"])
            except json.JSONDecodeError:
                print("Invalid JSON entry:", line)
    # with open(output_file_path, "w") as output_file:
    #      json.dump(extracted_requests, output_file, indent=4)
    #  # Load the JSON data
    # with open(input_json_path, 'r', encoding='utf-8') as file:
    #     json_data = json.load(file)
    
    # Normalize JSON data into a flat table
    df = pd.json_normalize(extracted_requests)
    
    # Combine all headers into a single dictionary under the 'headers' column
    df['headers'] = df[[col for col in df.columns if col.startswith('headers.')]].to_dict(orient='records')
    # Select only the 'request_line', 'headers', and 'body' columns
    df_selected = df[['request_line', 'headers', 'body']]
    
    # Save the selected columns to a CSV file
    df_selected.to_csv(output_file_path, index=False)
    
    print(f"CSV file created at: {output_file_path}")
```

```python
import json
import csv
import re
cnt=0
# Define the input log file path
file_path = 'example-xss.log'

# Define the output CSV file path
output_csv = 'example-xss.csv'

# Initialize a list to hold the results
results = []

# Regex pattern to find "paranoia-level/{a number}"
paranoia_pattern = re.compile(r'paranoia-level/(\d+)')

# Load and process the log file
with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
    log_data = file.readlines()

# Iterate over each log line
for line in log_data:
    try:
        # Convert the JSON string to a Python dictionary
        transaction = json.loads(line)
        
        # Access the transaction's messages (if available)
        messages = transaction.get('transaction', {}).get('messages', [])
        
        # Iterate over the messages to find SQL injection attacks
        for message in messages:
            details = message.get('details', {})
            file_text = details.get('file', '')
            
            # Check if the file contains "APPLICATION-ATTACK-SQLI"
            if "APPLICATION-ATTACK-XSS" in file_text:
                # Extract the necessary information
                request_url = transaction['transaction']['request'].get('uri', '')
                attack_data = details.get('data', '')
                
                # Extract paranoia level from the tags if available
                tags = details.get('tags', [])
                paranoia_level = 'N/A'
                
                # Check for the paranoia-level tag
                for tag in tags:
                    match = paranoia_pattern.search(tag)
                    if match:
                        paranoia_level = match.group(1)  # Extract the number
                        break
                
                # Add to results
                results.append([request_url, attack_data, paranoia_level])
                cnt+=1
                break
    except json.JSONDecodeError:
        print("# If there's an error in parsing, skip the line")
        continue

# Write the results to a CSV file
with open(output_csv, mode='w', newline='', encoding='utf-8', errors='ignore') as file:
    writer = csv.writer(file)
    writer.writerow(['Request URL', 'Attack Data', 'Paranoia Level'])  # Write header
    writer.writerows(results)  # Write the data
print(cnt)
print(f'Results saved to {output_csv}')
```
